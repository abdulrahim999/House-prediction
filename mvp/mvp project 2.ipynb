```python
# Create your feature matrix (X) and target vector (y)
y, X = patsy.dmatrices('price ~ bedrooms + bathrooms + sqft_living + grade + view', data=df, return_type="dataframe")
```

```python
# Create your model
model = sm.OLS(y, X)

# Fit your model to your training set
fit = model.fit()

# Print summary statistics of the model's performance
fit.summary()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=0)
```

```python
lr=LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_predict= lr.predict(X_test[:5])
y_predict
```

```python
train_data_score=lr.score(X_train,y_train)
test_data_score=lr.score(X_test,y_test)
print(train_data_score)
print(test_data_score)
```

```python
y_pred=lr.predict(X_test)
r2_score(y_test,y_pred)
```
